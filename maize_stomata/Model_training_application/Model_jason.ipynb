```python
sample_dir = "./samples_prediction/Oscar_resized/"

file_list = [file for file in os.listdir(sample_dir) if file.endswith('.png')]
# Get the full paths of PNG files in the sample directory
full_file_paths = [os.path.join(sample_dir, file) for file in file_list if file.endswith(".png")]

from PIL import Image
for file_name in file_list:
    file_path = os.path.join(sample_dir, file_name)
    with Image.open(file_path) as img:
        width, height = img.size
        #print(f"File: {file_name}, Size: {width}x{height}")
```

```python
import os
import cv2
import json

# Assuming predictor is already defined and set up

for file_path in full_file_paths:
    print(file_path)
    image = cv2.imread(file_path)
    outputs = predictor(image)
    filename = os.path.basename(file_path)
    genotype = filename.split('_')[1]

    # Extract relevant information from the outputs
    data = {
        "filename": filename,
        "objects": [],
        "genotype": genotype
    }

    for i in range(len(outputs["instances"])):
        class_id = outputs["instances"].pred_classes[i].item()
        mask = outputs["instances"].pred_masks[i].cpu().numpy().tolist()
        bbox = outputs["instances"].pred_boxes.tensor[i].cpu().numpy().tolist()
        score = outputs["instances"].scores[i].item()

        obj_data = {
            "class_name": class_id,
            "bbox": bbox,
            "mask": mask,
            "score": score
        }
        data["objects"].append(obj_data)

    # Write JSON to a file
    json_filename = filename.split('.')[0] + '.json'  # Change file extension to .json
    with open(json_filename, 'w') as json_file:
        json.dump(data, json_file)
```
